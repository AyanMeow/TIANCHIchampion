```python
import pandas as pd
```

```python
machine=pd.read_csv('./adaptData/machine-train.csv')
machine
```

```python
machine_dd=machine['text'].to_frame()
machine_dd['label']=1
machine_dd.to_csv('./adaptData/processed_machine.csv',index=None)
machine_dd
```

```python
human=pd.read_table('./adaptData/training_set_rel3.tsv',encoding='unicode_escape')
human=human.sample(n=3340)['essay'].to_frame().reset_index(drop=True)
human
```

```python
def process_dear(text:str):
    if text.startswith('Dear'):
        idx=text.find(',')
        text=text[idx+2:]
    text=text.replace('\n','').replace('\\','').replace('/','').replace('@','').replace('','')
    text=text.strip(' ')
    return text
```

```python
human_dd=human['essay'].map(process_dear).to_frame()
human_dd
```

```python
human_dd.columns=['text']
human_dd['label']=0
human_dd
```

```python
human_dd.to_csv('./adaptData/processed_human.csv',index=None)
```

```python
con=pd.concat([machine_dd,human_dd])
con=con.sample(frac=1).reset_index(drop=True)
con.to_csv('./adaptData/processed_mixed.csv',index=None)
con
```
