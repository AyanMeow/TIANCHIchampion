```python
def minDeletion(nums):
    n=len(nums)
    if n==1:
        return 1
    dp=[0]*(n+1)
    for i in range(n-1,-1,-1):
        if i % 2 == 0 and i+1 < n :
            if nums[i] != nums[i+1] :
                dp[i]=dp[i+1]+1
            else:
                dp[i]=dp[i+1]
        else:
            dp[i]=dp[i+1]+1
    
    if dp[0] % 2 != 0:
        return n - dp[0]-1
    else:
        return n-dp[0]
```

```python
def entityParser(text: str) -> str:
    entity_map = {
        "&quot;": "\"",
        "&apos;": "'",
        "&amp;": "&",
        "&gt;": ">",
        "&lt;": "<",
        "&frasl;": "/"
    }
    flag=True
    while flag:
        for key,value in entity_map.items():
            text=text.replace(key,value)
        flag=False
        for key,value in entity_map.items():
            if key in text:
                flag=True 
    return text
```

```python
def getk(root,idx,k,res,target):
    if k == 0:
        if root[idx] != None and root[idx] != target:
            res.append(root[idx])
    else:
        if idx*2 < len(root):
            getk(root,idx*2,k-1,res,target)
        if idx*2+1 < len(root):
            getk(root,idx*2+1,k-1,res,target)
        if idx//2 > 0:
            getk(root,idx//2,k-1,res,target)
```
