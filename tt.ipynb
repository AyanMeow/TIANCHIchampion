```python
from typing import List
import collections
import bisect
from typing import List
import collections
import bisect
class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        stack = []
        res,i = 0,0
        while i < len(nums):
            if len(stack)>0 and sum(stack) == k:
                res +=1
                stack.append(nums[i])
                i += 1
            elif len(stack)<=0 or sum(stack) < k:
                stack.append(nums[i])
                i += 1
            elif sum(stack) > k:
                while sum(stack) > k:
                    stack = stack[1:]
        while sum(stack) > k:
            stack = stack[1:]
        if len(stack)>0 and sum(stack) == k :res+=1
        return res
```

```python
def findPeakGrid(mat):
    m,n = len(mat),len(mat[0])
    pad_mat = []
    for i in range(m):
        pad_mat.append( [-1]+mat[i]+[-1])
    pad_mat.insert(0,[-1]*(n+2))
    pad_mat.append([-1]*(n+2))
    for i in range(1,m+1):
        low, high = 1, n
        while low < high:
            mid = (low+high)//2
            if pad_mat[i][mid] > pad_mat[i][mid+1] :
                high = mid
            else:
                low = mid + 1
        if pad_mat[i][low] > pad_mat[i-1][low] and pad_mat[i][low] > pad_mat[i+1][low]:
            return [i-1,low-1]
```

```python
import datetime
import time
def str2utc(date_string:str,format_str:str = None) -> (str,datetime.datetime):
    formatstr = format_str or '%Y-%m-%dT%H:%M:%SZ'
    
    now_stamp = time.time()
    local_time = datetime.datetime.fromtimestamp(now_stamp)
    utc_time = datetime.datetime.utcfromtimestamp(now_stamp)
    offset = local_time - utc_time
    
    if 'T' in date_string:
        str_time = datetime.datetime.strptime(date_string, '%Y-%m-%dT%H:%M:%SZ')
    elif ':' in date_string:
        str_time = datetime.datetime.strptime(date_string, "%Y-%m-%d-%H:%M")
    else:
        str_time = datetime.datetime.strptime(date_string, '%Y-%m-%d')
    # 将本地时间转换为UTC时间
    utc_time = str_time - offset
    return utc_time.strftime(formatstr),utc_time
```
